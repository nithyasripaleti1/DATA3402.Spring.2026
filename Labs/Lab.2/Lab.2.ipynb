````python
def draw_board(board):
    n = len(board)

    for i in range(n):
        print("---" * n)
        row = "|"
        for j in range(n):
            if board[i][j] == 0:
                row += "   |"
            elif board[i][j] == 1:
                row += " X |"
            else:
                row += " O |"
        print(row)

    print("---" * n)
````

*Exercise 4:* Write a function that takes a `n` by `n` matrix representing a tic-tac-toe game, and returns -1, 0, 1, or 2 indicating the game is incomplete, the game is a draw, player 1 has won, or player 2 has one, respectively. Here are some example inputs you can use to test your code:

````python
def game_status(board):
    n = len(board)

    #rows
    for i in range(n):
        if board[i][0] != 0:
            win = True
            for j in range(n):
                if board[i][j] != board[i][0]:
                    win = False
            if win:
                return board[i][0]

    #columns
    for j in range(n):
        if board[0][j] != 0:
            win = True
            for i in range(n):
                if board[i][j] != board[0][j]:
                    win = False
            if win:
                return board[0][j]

    #main diagonal
    if board[0][0] != 0:
        win = True
        for i in range(n):
            if board[i][i] != board[0][0]:
                win = False
        if win:
            return board[0][0]

    #other diag
    if board[0][n - 1] != 0:
        win = True
        for i in range(n):
            if board[i][n - 1 - i] != board[0][n - 1]:
                win = False
        if win:
            return board[0][n - 1]

    #empty spaces
    for row in board:
        if 0 in row:
            return -1

   
    return 0
````

````python
winner_is_2 = [[2, 2, 0],
	[2, 1, 0],
	[2, 1, 1]]

winner_is_1 = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 1]]

winner_is_also_1 = [[0, 1, 0],
	[2, 1, 0],
	[2, 1, 1]]

no_winner = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 2]]

also_no_winner = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 0]]

print(game_status(winner_is_2))        
print(game_status(winner_is_1))        
print(game_status(winner_is_also_1))   
print(game_status(no_winner))          
print(game_status(also_no_winner)) 
````

*Exercise 5:* Write a function that takes a game board, player number, and `(x,y)` coordinates and places "X" or "O" in the correct location of the game board. Make sure that you only allow filling previously empty locations. Return `True` or `False` to indicate successful placement of "X" or "O".

````python
def place_move(board, player, row, col):
    n = len(board)

    if row < 0 or row >= n or col < 0 or col >= n:
        return False

    if board[row][col] != 0:
        return False

    board[row][col] = player
    return True
````

*Exercise 6:* Modify Exercise 3 to show column and row labels so that players can specify location using "A2" or "C1".

````python
def draw_board_labeled(board):
    n = len(board)

    print("  ", end="")
    for i in range(n):
        print(chr(ord("A") + i), end="   ")
    print()

    for i in range(n):
        print(" ---" * n)
        print(i + 1, end="|")

        for j in range(n):
            if board[i][j] == 0:
                print("   |", end="")
            elif board[i][j] == 1:
                print(" X |", end="")
            else:
                print(" O |", end="")
        print()

    print(" ---" * n)
````

*Exercise 7:* Write a function that takes a board, player number, and location specified as in exercise 6 and then calls exercise 5 to correctly modify the board.  

````python
def place_move_location(board, player, loc):
    loc = loc.strip().upper()

    if len(loc) < 2:
        return False

    col_letter = loc[0]
    row_part = loc[1:]

    if not col_letter.isalpha():
        return False
    if not row_part.isdigit():
        return False

    col = ord(col_letter) - ord("A")
    row = int(row_part) - 1

    return place_move(board, player, row, col)
````

````python
b = make_board(3)
draw_board_labeled(b)

print(place_move_location(b, 1, "A1"))  
print(place_move_location(b, 2, "A1"))  

draw_board_labeled(b)
````

*Exercise 8:* Write a function is called with a board and player number, takes input from the player using python's `input`, and modifies the board using your function from exercise 7. Note that you should keep asking for input until you have gotten a valid input that results in a valid move.

````python
def ask_player_move(board, player):
    while True:
        loc = input("player " + str(player) + " enter move: ")
        ok = place_move_location(board, player, loc)

        if ok:
            return
        else:
            print("invalid move. Try again.")
````

*Exercise 9:* Use all of the previous exercises to implement a full tic-tac-toe game, where an appropriate board is drawn, 2 players are repeatedly asked for a location coordinates of where they wish to place a mark, and the game status is checked until a player wins or a draw occurs.

````python
def play_game(n):
    board = make_board(n)
    player = 1

    while True:
        draw_board_labeled(board)
        ask_player_move(board, player)

        status = game_status(board)

        if status == 1:
            draw_board_labeled(board)
            print("Player 1 wins!")
            break
        elif status == 2:
            draw_board_labeled(board)
            print("Player 2 wins!")
            break
        elif status == 0:
            draw_board_labeled(board)
            print("Draw!")
            break

        if player == 1:
            player = 2
        else:
            player = 1
````

Exercise 11: Develop a version of the game where one player is the computer. Note that you don't need to do an extensive seach for the best move. You can have the computer simply protect against loosing and otherwise try to win with straight or diagonal patterns.

````python
def get_empty_spots(board):
    spots = []
    n = len(board)

    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                spots.append((r, c))

    return spots
````

````python
def choose_computer_move(board, computer_player):
    opponent = 1
    if computer_player == 1:
        opponent = 2

    spots = get_empty_spots(board)

    # 1. to win
    for (r, c) in spots:
        board[r][c] = computer_player
        if game_status(board) == computer_player:
            board[r][c] = 0
            return (r, c)
        board[r][c] = 0

    #2. to block
    for (r, c) in spots:
        board[r][c] = opponent
        if game_status(board) == opponent:
            board[r][c] = 0
            return (r, c)
        board[r][c] = 0

   
    if len(spots) > 0:
        return spots[0]

    return None
````
